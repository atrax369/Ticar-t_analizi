## ilə görə ən çox satışı gördük

```python
sales_by_year = df.groupby('order_year').agg({'customer_value_total_ever_offline': 'sum', 'customer_value_total_ever_online': 'sum'}).reset_index()
```

```python
print(sales_by_year)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='order_year', y='customer_value_total_ever_offline', data=sales_by_year, color='blue', label='Offline Sales')
sns.barplot(x='order_year', y='customer_value_total_ever_online', data=sales_by_year, color='orange', label='Online Sales', alpha=0.7)
plt.title('Total Sales by Year')
```

```python
df['ferq']=df[['order_num_total_ever_offline','order_num_total_ever_online']].sum()  
```

```python
online_toplam=df['order_num_total_ever_online'].sum()
offline_toplam=df['order_num_total_ever_offline'].sum()
```

```python
labels = ['Online Sales', 'Offline Sales']
sizes = [online_toplam, offline_toplam]     
colors = ['orange', 'blue']
explode = (0.1, 0)  
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Distribution of Online and Offline Sales')   
```

```python
df.head()
```

```python
df['total_sales'] = df['customer_value_total_ever_offline'] + df['customer_value_total_ever_online']
```

```python
mean_total_sales = df['total_sales'].mean()
```

```python
vip_customers = df[df['total_sales'] > mean_total_sales]
normal_customers = df[df['total_sales'] <= mean_total_sales]
print("VIP Customers:")
print(len(vip_customers)) 
print("Normal Customers:")
print(len(normal_customers))   
```

```python
vip_customers['order_channel'].value_counts().plot(kind='bar', color=['orange', 'blue', 'green', 'red'])
plt.title('Order Channel Distribution for VIP Customers')
```

```python
normal_customers['order_channel'].value_counts().plot(kind='bar', color=['orange', 'blue', 'green', 'red'])
plt.title('Order Channel Distribution for Normal Customers')
```
